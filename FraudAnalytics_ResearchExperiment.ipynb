6. Observations

- A short summary of what was discovered. Suspicious trends, high-r variables, or visual insights

```python
# Summary of observations

def summarize_correlations(correlation_matrix, threshold=0.5):
    """
    Summarize high correlations in the dataset
    
    Parameters:
    correlation_matrix: pandas DataFrame of correlations
    threshold: absolute correlation value to consider as significant
    """
    # Find high positive correlations (excluding self-correlations)
    high_pos_corr = []
    
    # Find high negative correlations
    high_neg_corr = []
    
    # Iterate through the correlation matrix
    for i in range(len(correlation_matrix.columns)):
        for j in range(i+1, len(correlation_matrix.columns)):
            corr_value = correlation_matrix.iloc[i, j]
            feature_pair = (correlation_matrix.columns[i], correlation_matrix.columns[j])
            
            if corr_value > threshold:
                high_pos_corr.append((feature_pair, corr_value))
            elif corr_value < -threshold:
                high_neg_corr.append((feature_pair, corr_value))
    
    # Sort by correlation strength
    high_pos_corr.sort(key=lambda x: x[1], reverse=True)
    high_neg_corr.sort(key=lambda x: x[1])
    
    print("=== CORRELATION ANALYSIS SUMMARY ===\n")
    
    # Print high positive correlations
    print(f"Strong Positive Correlations (r > {threshold}):")
    if high_pos_corr:
        for (feat1, feat2), corr in high_pos_corr:
            print(f"  • {feat1} and {feat2}: {corr:.3f}")
    else:
        print("  • None found")
    
    # Print high negative correlations
    print(f"\nStrong Negative Correlations (r < -{threshold}):")
    if high_neg_corr:
        for (feat1, feat2), corr in high_neg_corr:
            print(f"  • {feat1} and {feat2}: {corr:.3f}")
    else:
        print("  • None found")
    
    # Check for multicollinearity concerns
    if high_pos_corr:
        print("\nPotential Multicollinearity Concerns:")
        for (feat1, feat2), corr in high_pos_corr:
            if corr > 0.8:
                print(f"  • {feat1} and {feat2} (r={corr:.3f}): Consider removing one of these features")
    
    # Check correlation with target variable (if 'Class' exists)
    if 'Class' in correlation_matrix.columns:
        print("\nCorrelation with Target Variable (Class):")
        class_corr = [(col, correlation_matrix.loc['Class', col]) 
                      for col in correlation_matrix.columns if col != 'Class']
        class_corr.sort(key=lambda x: abs(x[1]), reverse=True)
        
        for feat, corr in class_corr[:10]:  # Top 10 correlations
            print(f"  • {feat}: {corr:.3f}")
    
    print("\n=== HERE ARE MY OBSERVATIONS BELOW IN THE NEXT CELL===")
    print("""

""")

# Summary function
summarize_correlations(correlation_matrix)
```

```python
# Plotting the fraud rate over time using the Time column in the credit card dataset (creditcard2.csv)

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# df is already loaded
# df = pd.read_csv("creditcard2.csv")

# Convert Time to hours for better readability (Time is in seconds from the first transaction)
df['Hour'] = df['Time'] / 3600  # Convert seconds to hours

# Group data by hour and calculate fraud rate
# First, create hour bins (e.g., every 1 hour)
hour_bins = np.linspace(df['Hour'].min(), df['Hour'].max(), 24)  # 24 bins across the time range
df['HourBin'] = pd.cut(df['Hour'], bins=hour_bins)

# Calculate fraud rate per hour bin
fraud_by_hour = df.groupby('HourBin').agg(
    fraud_count=('Class', lambda x: (x == 1).sum()),
    total_count=('Class', 'count')
)
fraud_by_hour['fraud_rate'] = fraud_by_hour['fraud_count'] / fraud_by_hour['total_count'] * 100

# Create the plot
plt.figure(figsize=(14, 7))

# Plot fraud rate
ax1 = plt.gca()
ax1.set_xlabel('Time (hours)', fontsize=12)
ax1.set_ylabel('Fraud Rate (%)', fontsize=12, color='blue')
ax1.plot(fraud_by_hour.index.astype(str), fraud_by_hour['fraud_rate'], 'o-', color='blue', linewidth=2)
ax1.tick_params(axis='y', labelcolor='blue')
ax1.grid(True, linestyle='--', alpha=0.7)

# Add transaction volume as a secondary axis
ax2 = ax1.twinx()
ax2.set_ylabel('Transaction Count', fontsize=12, color='red')
ax2.bar(fraud_by_hour.index.astype(str), fraud_by_hour['total_count'], alpha=0.2, color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Add title and adjust layout
plt.title('Fraud Rate and Transaction Volume Over Time', fontsize=16)
plt.tight_layout()

# Add legend
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, ['Fraud Rate', 'Transaction Count'], loc='upper left')

plt.show()

# Print some statistics
print(f"Average fraud rate: {fraud_by_hour['fraud_rate'].mean():.2f}%")
print(f"Maximum fraud rate: {fraud_by_hour['fraud_rate'].max():.2f}% (at {fraud_by_hour['fraud_rate'].idxmax()})")
```

```python
# Use `groupby()` to explore fraud by amount bins 


import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# df is already loaded
# df = pd.read_csv("creditcard2.csv")

# Create amount bins
# Define bin edges - adjust these based on your data distribution
amount_bins = [0, 10, 50, 100, 500, 1000, 5000, df['Amount'].max() + 1]
bin_labels = ['0-10', '10-50', '50-100', '100-500', '500-1000', '1000-5000', '5000+']

# Create a new column with binned amounts
df['AmountBin'] = pd.cut(df['Amount'], bins=amount_bins, labels=bin_labels)

# Group by amount bins and calculate fraud statistics
fraud_by_amount = df.groupby('AmountBin').agg(
    total_transactions=('Class', 'count'),
    fraud_count=('Class', sum),
    avg_amount=('Amount', 'mean')
)

# Calculate fraud rate
fraud_by_amount['fraud_rate'] = fraud_by_amount['fraud_count'] / fraud_by_amount['total_transactions'] * 100

# Display the results
print("Fraud Analysis by Transaction Amount")
print("===================================")
print(fraud_by_amount)

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))

# Plot 1: Fraud rate by amount bin
fraud_by_amount['fraud_rate'].plot(kind='bar', color='darkred', ax=ax1)
ax1.set_title('Fraud Rate by Transaction Amount', fontsize=14)
ax1.set_ylabel('Fraud Rate (%)', fontsize=12)
ax1.set_xlabel('Transaction Amount Range ($)', fontsize=12)
ax1.grid(axis='y', linestyle='--', alpha=0.7)

# Add value labels on top of bars
for i, v in enumerate(fraud_by_amount['fraud_rate']):
    ax1.text(i, v + 0.1, f'{v:.2f}%', ha='center', fontsize=10)

# Plot 2: Transaction volume by amount bin with fraud breakdown
fraud_by_amount[['fraud_count', 'total_transactions']].plot(
    kind='bar', 
    ax=ax2,
    color=['darkred', 'lightgray'],
    stacked=False
)
ax2.set_title('Transaction Count by Amount Range', fontsize=14)
ax2.set_ylabel('Number of Transactions', fontsize=12)
ax2.set_xlabel('Transaction Amount Range ($)', fontsize=12)
ax2.grid(axis='y', linestyle='--', alpha=0.7)
ax2.legend(['Fraudulent', 'Total'])

# Add a text box with key observations
textstr = '\n'.join((
    'Key Observations:',
    f'- Highest fraud rate: {fraud_by_amount["fraud_rate"].max():.2f}% in {fraud_by_amount["fraud_rate"].idxmax()} range',
    f'- Lowest fraud rate: {fraud_by_amount["fraud_rate"].min():.2f}% in {fraud_by_amount["fraud_rate"].idxmin()} range',
    f'- Most fraudulent transactions: {fraud_by_amount["fraud_count"].max()} in {fraud_by_amount["fraud_count"].idxmax()} range'
))
props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
ax2.text(0.05, 0.95, textstr, transform=ax2.transAxes, fontsize=10,
        verticalalignment='top', bbox=props)

plt.tight_layout()
plt.show()

# Create a heatmap showing fraud rate by amount bin and time of day
# First, create time of day bins
df['Hour'] = (df['Time'] / 3600) % 24  # Convert seconds to hours of day (0-24)
df['TimeOfDay'] = pd.cut(
    df['Hour'], 
    bins=[0, 6, 12, 18, 24], 
    labels=['Night (0-6)', 'Morning (6-12)', 'Afternoon (12-18)', 'Evening (18-24)']
)

# Create a pivot table for the heatmap
fraud_heatmap = pd.pivot_table(
    df, 
    values='Class',
    index='AmountBin',
    columns='TimeOfDay',
    aggfunc=lambda x: 100 * sum(x) / len(x)  # Fraud rate as percentage
)

plt.figure(figsize=(10, 8))
sns.heatmap(fraud_heatmap, annot=True, cmap='YlOrRd', fmt='.2f', linewidths=.5)
plt.title('Fraud Rate (%) by Amount Range and Time of Day', fontsize=14)
plt.tight_layout()
plt.show()
```

```python

# Pairplot with Seaborn for high-correlation variables

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# df is already loaded
# df = pd.read_csv("creditcard2.csv")

# Calculate the correlation matrix
correlation_matrix = df.select_dtypes(include=[np.number]).corr()

# Find features with high correlation to the Class variable (fraud indicator)
# or high correlation with each other
def get_high_correlation_features(corr_matrix, threshold=0.5, target_col='Class', max_features=6):
    """
    Get features that have high correlation with the target or with each other
    
    Parameters:
    corr_matrix: correlation matrix
    threshold: minimum absolute correlation value to consider
    target_col: target column name (usually 'Class' for fraud detection)
    max_features: maximum number of features to return
    
    Returns:
    list of feature names
    """
    # If target column exists, prioritize features correlated with it
    high_corr_features = []
    
    if target_col in corr_matrix.columns:
        # Get correlations with target, sorted by absolute value
        target_corrs = corr_matrix[target_col].drop(target_col)
        high_target_corrs = target_corrs[abs(target_corrs) > threshold].sort_values(key=abs, ascending=False)
        high_corr_features.extend(high_target_corrs.index.tolist())
    
    # Add features with high correlation between them
    for i in range(len(corr_matrix.columns)):
        for j in range(i+1, len(corr_matrix.columns)):
            col1, col2 = corr_matrix.columns[i], corr_matrix.columns[j]
            if col1 != target_col and col2 != target_col:  # Skip target column
                corr_value = abs(corr_matrix.iloc[i, j])
                if corr_value > threshold:
                    if col1 not in high_corr_features:
                        high_corr_features.append(col1)
                    if col2 not in high_corr_features:
                        high_corr_features.append(col2)
    
    # Always include the target column if it exists
    if target_col in corr_matrix.columns and target_col not in high_corr_features:
        high_corr_features.append(target_col)
    
    # Limit the number of features to avoid overcrowded plot
    return high_corr_features[:max_features]

# Get features with high correlation
high_corr_features = get_high_correlation_features(
    correlation_matrix, 
    threshold=0.5,  # Adjust this threshold as needed
    max_features=6  # Adjust based on how many features you want to include
)

# Add Class column if it exists and not already included
if 'Class' in df.columns and 'Class' not in high_corr_features:
    high_corr_features.append('Class')

# Print selected features
print("Selected features for pairplot:")
print(high_corr_features)

# Create a sample if the dataset is large (optional)
# This helps make the pairplot generation faster
sample_size = min(10000, len(df))  # Adjust sample size as needed
df_sample = df.sample(sample_size, random_state=42)

# Create a new column to color points by fraud/non-fraud
if 'Class' in df_sample.columns:
    df_sample['Fraud'] = df_sample['Class'].map({0: 'Non-Fraud', 1: 'Fraud'})
    hue_column = 'Fraud'
else:
    hue_column = None

# Create the pairplot
plt.figure(figsize=(12, 10))
pairplot = sns.pairplot(
    df_sample[high_corr_features + (['Fraud'] if hue_column == 'Fraud' else [])], 
    hue=hue_column,
    corner=True,  # Show only the lower triangle to avoid redundancy
    plot_kws={'alpha': 0.5, 's': 10},  # Make points semi-transparent and smaller
#    diag_kws={'bins': 20},  # Number of bins in histogram
    palette={'Fraud': 'red', 'Non-Fraud': 'blue'} if hue_column == 'Fraud' else None
)

# Add title
pairplot.fig.suptitle('Pairplot of Highly Correlated Features', y=1.02, fontsize=16)

# Add correlation coefficients to the scatter plots
for i, j in zip(*np.triu_indices_from(pairplot.axes, 1)):
    if i < len(high_corr_features) and j < len(high_corr_features):
        feature_i = high_corr_features[i]
        feature_j = high_corr_features[j]
        corr = df[feature_i].corr(df[feature_j])
        pairplot.axes[j, i].annotate(
            f'r = {corr:.2f}',
            xy=(0.5, 0.9),
            xycoords='axes fraction',
            ha='center',
            fontsize=9,
            bbox=dict(boxstyle='round,pad=0.3', fc='white', alpha=0.7)
        )

plt.tight_layout()
plt.show()

# Optional: Print correlation matrix for selected features
print("\nCorrelation matrix for selected features:")
print(correlation_matrix.loc[high_corr_features, high_corr_features].round(2))

```

```python
# Z-score or IQR for outlier detection

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Function to detect outliers using Z-score method
def detect_outliers_zscore(df, features, threshold=3):
    """
    Detect outliers using Z-score method
    
    Parameters:
    df (DataFrame): Input DataFrame
    features (list): List of features to check for outliers
    threshold (float): Z-score threshold (default: 3)
    
    Returns:
    DataFrame: DataFrame with outlier flags
    """
    outlier_indices = {}
    outlier_df = df.copy()
    
    for feature in features:
        # Skip non-numeric columns
        if not pd.api.types.is_numeric_dtype(df[feature]):
            continue
            
        # Calculate z-scores
        z_scores = np.abs(stats.zscore(df[feature], nan_policy='omit'))
        
        # Find outliers
        outliers = np.where(z_scores > threshold)[0]
        outlier_indices[feature] = outliers
        
        # Create outlier flag column
        outlier_df[f'{feature}_zscore_outlier'] = (z_scores > threshold).astype(int)
    
    return outlier_df, outlier_indices

# Function to detect outliers using IQR method
def detect_outliers_iqr(df, features, multiplier=1.5):
    """
    Detect outliers using IQR method
    
    Parameters:
    df (DataFrame): Input DataFrame
    features (list): List of features to check for outliers
    multiplier (float): IQR multiplier (default: 1.5)
    
    Returns:
    DataFrame: DataFrame with outlier flags
    """
    outlier_indices = {}
    outlier_df = df.copy()
    
    for feature in features:
        # Skip non-numeric columns
        if not pd.api.types.is_numeric_dtype(df[feature]):
            continue
            
        # Calculate Q1, Q3 and IQR
        Q1 = df[feature].quantile(0.25)
        Q3 = df[feature].quantile(0.75)
        IQR = Q3 - Q1
        
        # Define bounds
        lower_bound = Q1 - multiplier * IQR
        upper_bound = Q3 + multiplier * IQR
        
        # Find outliers
        outliers = df[(df[feature] < lower_bound) | (df[feature] > upper_bound)].index
        outlier_indices[feature] = outliers
        
        # Create outlier flag column
        outlier_df[f'{feature}_iqr_outlier'] = ((df[feature] < lower_bound) | 
                                               (df[feature] > upper_bound)).astype(int)
    
    return outlier_df, outlier_indices

# Apply both methods to the selected features
outlier_df_zscore, outlier_indices_zscore = detect_outliers_zscore(df, high_corr_features)
outlier_df_iqr, outlier_indices_iqr = detect_outliers_iqr(df, high_corr_features)

# Print summary of outliers detected
print("Outlier Detection Summary:")
print("\nZ-score method (threshold = 3):")
for feature in high_corr_features:
    if feature in outlier_indices_zscore:
        count = len(outlier_indices_zscore[feature])
        percentage = (count / len(df)) * 100
        print(f"{feature}: {count} outliers ({percentage:.2f}%)")

print("\nIQR method (multiplier = 1.5):")
for feature in high_corr_features:
    if feature in outlier_indices_iqr:
        count = len(outlier_indices_iqr[feature])
        percentage = (count / len(df)) * 100
        print(f"{feature}: {count} outliers ({percentage:.2f}%)")

# Visualize outliers for a few selected features
# Create a sample if the dataset is large
sample_size = min(10000, len(df))
df_sample = df.sample(sample_size, random_state=42)

# Plot boxplots to visualize outliers
plt.figure(figsize=(15, len(high_corr_features) * 4))
for i, feature in enumerate(high_corr_features):
    if not pd.api.types.is_numeric_dtype(df[feature]):
        continue
        
    plt.subplot(len(high_corr_features), 2, 2*i+1)
    sns.boxplot(x=df_sample[feature])
    plt.title(f'Boxplot of {feature}')
    plt.tight_layout()
    
    plt.subplot(len(high_corr_features), 2, 2*i+2)
    plt.scatter(range(len(df_sample)), df_sample[feature], alpha=0.5)
    
    # Add horizontal lines for Z-score thresholds
    mean = df_sample[feature].mean()
    std = df_sample[feature].std()
    plt.axhline(y=mean + 3*std, color='r', linestyle='--', label='Z-score = 3')
    plt.axhline(y=mean - 3*std, color='r', linestyle='--')
    
    # Add horizontal lines for IQR thresholds
    Q1 = df_sample[feature].quantile(0.25)
    Q3 = df_sample[feature].quantile(0.75)
    IQR = Q3 - Q1
    plt.axhline(y=Q3 + 1.5*IQR, color='g', linestyle='--', label='IQR = 1.5')
    plt.axhline(y=Q1 - 1.5*IQR, color='g', linestyle='--')
    
    plt.title(f'Scatter plot of {feature} with outlier thresholds')
    plt.legend()
    plt.tight_layout()

plt.tight_layout()
plt.show()

# Create a correlation heatmap of outlier flags
# This can help identify if outliers tend to occur in the same observations across features
outlier_cols = [col for col in outlier_df_zscore.columns if '_outlier' in col]
if len(outlier_cols) > 1:  # Only create heatmap if we have multiple outlier columns
    plt.figure(figsize=(12, 10))
    sns.heatmap(outlier_df_zscore[outlier_cols].corr(), annot=True, cmap='coolwarm', fmt='.2f')
    plt.title('Correlation between Outlier Flags')
    plt.tight_layout()
    plt.show()

# Count observations with multiple outliers
outlier_counts = outlier_df_zscore[outlier_cols].sum(axis=1)
print("\nNumber of features with outliers per observation (Z-score method):")
print(outlier_counts.value_counts().sort_index())

# If 'Class' exists, check if outliers are more common in one class
if 'Class' in df.columns:
    print("\nOutlier distribution by class:")
    for feature in high_corr_features:
        if feature in outlier_indices_zscore:
            outlier_class = df.loc[outlier_indices_zscore[feature], 'Class'].value_counts(normalize=True)
            print(f"\n{feature} outliers by class (Z-score):")
            print(outlier_class)


```
